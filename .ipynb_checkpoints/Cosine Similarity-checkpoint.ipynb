```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('all_data.csv')
```

```python
df = df[['team_number','OPR', 'CCWM', 'ClimbPoints', 'AutoPoints', 'OwnershipPoints', 'VaultPoints' ]]
```

```python
df.head(20)
```

```python
cols = list(df)
test = df[cols[1:]]
team_name = df[cols[0]]
test.head(20)
```

```python
from sklearn import preprocessing
x = test.values #returns a numpy array
print x.shape
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
df.head()
team_name = pd.DataFrame(team_name)
team_name.shape
result = pd.concat([team_name, df], axis=1)
result = result.head(50)
```

```python
result = pd.concat([team_name, df], axis=1)
result.head()
```

```python
# result = df.head(50)
result = result.head(50)
cos = euclidean_distances(result)
```

```python
cos[9]
# result = df
x_normed = cos / cos.max(axis=0)
print x_normed.shape
print x_normed
```

```python
_t = np.array(x_normed)
final_set = pd.DataFrame(_t, columns=list(result['team_number']), index=list(result['team_number']))

sns.set(font_scale=0.75)
plt.figure(figsize=(50,50))
colormap = plt.cm.cubehelix_r
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
print final_set.head()
sns.heatmap(final_set, annot=True, cmap=colormap, square=True, linewidths=0.2)
plt.title("Team Similarity")
plt.savefig('Team_Similarity.png')
plt.show()
```
